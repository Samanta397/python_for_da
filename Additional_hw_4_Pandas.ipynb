```python
from datetime import datetime

def transform_to_datetime64(date):
    if len(date) == 19:  # Формат "11.30.2022 10:26:37"
        return datetime.strptime(date, "%m.%d.%Y %H:%M:%S")
    elif len(date) == 16:  # Формат "2022/11/30 10:26"
        return datetime.strptime(date, "%m.%d.%Y %H:%M")

df["Timestamp"] = df.Timestamp.apply(lambda x: transform_to_datetime64(x))
df
```

```python
df['Знання англійської мови'] = pd.Categorical(df['Знання англійської мови'],
                      categories=['Elementary','Pre-Intermediate','Intermediate','Upper-Intermediate','Advanced'],
                      ordered=True)

df['Ваш тайтл'] = pd.Categorical(df['Ваш тайтл'],
                      categories=['Intern/Trainee', 'Junior', 'Немає тайтлу', 'Middle', 'Senior', 'Tech Lead', 'Team Lead', 'Architect', 'Head', 'Manager'],
                      ordered=True)
```

12. Зробіть зводну таблицю, де стовпці - це 'Ваш тайтл', рядки - 'Знання англійської мови', а значення - медіанна ЗП. Які висновки можна зробити?
Тут ще для зручності можна зробити градієнтне забарвлення і відформатувати значення для легшого перегляду.

```python
pivot_table = df.pivot_table(
    values="Зарплата у $$$ за місяць, лише ставка після сплати податків",
    index="Знання англійської мови",
    columns="Ваш тайтл",
    aggfunc="median"
)
pivot_table
```

Розглядаючі отримані дані, можна зробити висновок, що для позиції Інтерн рівень володіння англійскою не має особливовго значення, видно що зп коливається не залежно від рівня англійської. Це може бути показником того, що на даній позиції більшу увагу віддають технічним навичкам.

13. Виведіть результат методу describe() по колонцы розміру ЗП для топ 5 за частотою зустрічання в даних областей (В якій області ви зараз живете?). Зробіть висновки, де вигідніше жити з точки зору ЗП і чи суттєва різниця на Ваш погляд для прийняття рішення про переїзд?

14. Як зростає зарплата з роками? Для цього нам треба відформатувати колонку "Загальний стаж роботи за спеціальністю" аналогічно до того, як ми це робили для колонки 'Знання англійської мови' і потім вивести медіанні значення для кількості років стажу.

```python
def convert_experience(experience_str):
    if 'років' in experience_str or 'рік' in experience_str or 'роки' in experience_str:
        return int(experience_str.split()[0])  # Витягуємо число років
    elif 'Пів року' in experience_str:
        return 0.5  # Пів року зараховуємо як 0.5
    else:
        return 0

df["Загальний стаж роботи за спеціальністю"] = df["Загальний стаж роботи за спеціальністю"].apply(convert_experience)
median_salary_by_experience = df.groupby("Загальний стаж роботи за спеціальністю")["Зарплата у $$$ за місяць, лише ставка після сплати податків"].median()
median_salary_by_experience
```

Зарплата зростає в середньому десь на 300-500 дол за рік. При цьому можна побачити, що найбільші різніці в зп спостерігаються перші 4 роки роботи.
