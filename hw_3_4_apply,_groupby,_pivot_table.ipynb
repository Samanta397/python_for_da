1. Гіпотези
   1.1 Які товари та в якому місті купуються найбільше?
   1.2 Хто та які товари купує найбільше (чоловіки / жінки)?

```python
# 2.1 Які товари та в якому місті купуються найбільше?

#df.groupby(by=['Product line','City'])['Invoice ID'].count()

grouped = pd.pivot_table(df, index='City', columns='Product line', values='Invoice ID', aggfunc='count')
display(grouped)

max_values = grouped.max()
max_cities = grouped.idxmax()
result = pd.DataFrame({"Max Value": max_values, "City": max_cities})

print(result)
```

За результатами отриманих даних можна зробити висновки які групи товарів та в якому місті купуються найкраще і відповідно до них будувати маркетингову стратегію.

За результатами видно, що 
- в місті Naypyitaw краще продаються Fashion accessories та Food and beverages
- в місті Yangon краще продаються Electronic accessories та Home and lifestyle
- в місті Mandalay краще продаються Health and beauty та Sports and travel

```python
# 2.2 Хто та які товари купує найбільше (чоловіки / жінки)?

grouped = pd.pivot_table(df, index='Gender', columns='Product line', values='Invoice ID', aggfunc='count')
display(grouped)

max_values = grouped.max()
max_cities = grouped.idxmax()
result = pd.DataFrame({"Max Value": max_values, "Gender": max_cities})

print(result)
```

За результатами отриманих даних видно що 
- жінки більше купують товари Fashion accessories, Food and beverages та Sports and travel
- чоловіків більше цікавлять тьовари з груп Electronic accessories, Health and beauty та Home and lifestyle.

За допомогою цих даних можна розробляти рекламу розуміючи цільову категорію.

```python
# 2.3 Які групи товарів та в який час купуються? 

def get_hour(time_value):
    hour = time_value.split(':')[0]
    return int(hour)

df["Hour"] = df.Time.apply(lambda x: get_hour(x))
df["Hour"].value_counts()

grouped = pd.pivot_table(df, index='Hour', columns='Product line', values='Invoice ID', aggfunc='count')
display(grouped)

max_values = grouped.max()
max_cities = grouped.idxmax()
result = pd.DataFrame({"Max Value": max_values, "Hour": max_cities})

print(result)
```

На основі отриманих результатів можна бачити в які години краще продаються ті чи інші товари та відповідно робити різні акції типу "щасливих годин" для підвищення попиту а відповідно і продажів.
